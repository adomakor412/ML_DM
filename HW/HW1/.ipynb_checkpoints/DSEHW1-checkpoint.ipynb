```python
from DSEHW import DSELinearClassifier
import numpy as np
import numpy.random as nr
from sklearn.datasets import make_blobs
import pandas as pd
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt
from scipy.stats import norm

%matplotlib inline
```

# Update Rules

## Hyperbolic Tan (derivative)
### $$[\frac{2  e^{2x}}{1+e^{2x}}]^2$$
## Logistic Regression (derivative)
### $$\eta*\sum(y^{(i)}-w_i*x_i)$$

```python
nr.seed = 42
centers = [[1, 1], [-1, -1]]
X, y = make_blobs(n_samples=500, n_features=2, 
        centers=centers, cluster_std=1.0/8, center_box=(-10.0, 10.0), shuffle=True, random_state=nr.seed)
```

```python
df = pd.DataFrame({'x0':X.T[0], 
                   'x1':X.T[1],
                   'y': y,
                   'target':(np.array(['Blue','Red'])[y])})
```

```python
fig, ax = plt.subplots(1,figsize=(10,6))
_=sns.scatterplot(x='x0',y='x1',hue='target', data=df,ax=ax,alpha=.5,s=100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import SGDClassifier, Perceptron
from sklearn.linear_model import LogisticRegression

heldout = [0.95, 0.90, 0.75, 0.50, 0.01]
rounds = 20

X, y = make_blobs(n_samples=500, n_features=2, 
        centers=centers, cluster_std=1.0/8, center_box=(-10.0, 10.0), shuffle=True, random_state=nr.seed)

classifiers = [
    ("SGD", SGDClassifier(max_iter=100)),
    ("Perceptron", Perceptron()),
    ("SAG-LogisticReg", LogisticRegression(solver='sag', tol=1e-1, C=1.e4 / X.shape[0]))
]

xx = 1. - np.array(heldout)

for name, clf in classifiers:
    print("training %s" % name)
    rng = np.random.RandomState(42)
    yy = []
    for i in heldout:
        yy_ = []
        for r in range(rounds):
            X_train, X_test, y_train, y_test = \
                train_test_split(X, y, test_size=i, random_state=rng)
            clf.fit(X_train, y_train)
            y_pred = clf.predict(X_test)
            yy_.append(1 - np.mean(y_pred == y_test))
        yy.append(np.mean(yy_))
    plt.plot(xx, yy, label=name,alpha=.5)

plt.legend(loc="upper right")
plt.xlabel("Proportion train")
plt.ylabel("Test Error Rate")
#plt.scatter(xx,yy)
```

```python
for rate in [0.1,0.2,0.4]:
    heldout = [0.95, 0.90, 0.75, 0.50, 0.01]
    rounds = 20

    X, y = make_blobs(n_samples=500, n_features=2, 
            centers=centers, cluster_std=1.0/8, center_box=(-10.0, 10.0), shuffle=True, random_state=nr.seed)

    classifiers = [
        (f"SGD_{rate}", SGDClassifier(max_iter=100,tol=rate)),
        (f"Perceptron_{rate}", Perceptron(tol=rate)),
        (f"SAG-LogisticReg_{rate}", LogisticRegression(solver='sag', tol=rate, C=1.e4 / X.shape[0]))
    ]

    xx = 1. - np.array(heldout)

    for name, clf in classifiers:
        print("training %s" % name)
        rng = np.random.RandomState(42)
        yy = []
        for i in heldout:
            yy_ = []
            for r in range(rounds):
                X_train, X_test, y_train, y_test = \
                    train_test_split(X, y, test_size=i, random_state=rng)
                clf.fit(X_train, y_train)
                y_pred = clf.predict(X_test)
                yy_.append(1 - np.mean(y_pred == y_test))
            yy.append(np.mean(yy_))
        plt.plot(xx, yy, label=name,alpha=.75)

    plt.legend(loc="upper right")
    plt.xlabel("Proportion train")
    plt.ylabel("Test Error Rate")
    #plt.scatter(xx,yy)
```

# From Simple Gradient Descent, we can see that 0.1 is the best learning rate
## Apply rate of 0.1 to a sample_weight of (5,5) and plot. We are testing how well the models can train given an extreme deviance from the initial condition of "1" (1,1) because we have a 2D array.

```python
X, y = make_blobs(n_samples=500, n_features=2, 
            centers=centers, cluster_std=1.0/8, center_box=(-10.0, 10.0), shuffle=True, random_state=nr.seed)

for rate in [0.1]:
    heldout = [0.95, 0.90, 0.75, 0.50, 0.01]
    rounds = 20

    classifiers = [
        (f"SGD_{rate}", SGDClassifier(max_iter=100,tol=rate)),
        (f"Perceptron_{rate}", Perceptron(tol=rate)),
        (f"SAG-LogisticReg_{rate}", LogisticRegression(solver='sag', tol=rate, C=1.e4 / X.shape[0]))
    ]

    xx = 1. - np.array(heldout)

    for name, clf in classifiers:
        print("training %s" % name)
        rng = np.random.RandomState(42)
        yy = []
        for i in heldout:
            yy_ = []
            for r in range(rounds):
                X_train, X_test, y_train, y_test = \
                    train_test_split(X, y, test_size=i, random_state=rng)
                clf.fit(X_train, y_train, sample_weight = float(5))
                y_pred = clf.predict(X_test)
                yy_.append(1 - np.mean(y_pred == y_test))
            yy.append(np.mean(yy_))
        plt.plot(xx, yy, label=name,alpha=.75)

    plt.legend(loc="upper right")
    plt.xlabel("Proportion train")
    plt.ylabel("Test Error Rate")
    #plt.scatter(xx,yy)
```

## It's clear that from our models, as long as we choose a training set of at least 50% of the sample we are able to learn quickly, reducing the error to 0%

# Question 1


```python
df = pd.read_csv('http://bit.ly/2Cxfbe4')
```

```python
df = df.dropna()
df
```

```python
df2 = df['band_type']
df2 = pd.get_dummies(df2)
df['band_type']=df2['band']
df
```

```python
df = df.drop(columns='cylinder_size')
df = df.drop(columns='Unnamed: 0')
```

```python
df #Finally!
```

```python
df[df.columns[-1]]
```

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
```

```python
X = df[df.columns[:-1]]
y = df[df.columns[-1]].T

# Create the RFE object and rank each pixel
svc = SVC(kernel="linear", C=1)
rfe = RFE(estimator=svc, n_features_to_select=1, step=1)
rfe.fit(X, y)
```

## Ranking

```python
rfe.ranking_
```

```python
top_RFE = rfe.ranking_[:5]
```

```python
ind = list(top_RFE)
```

### These are your top features using RFE-SVM:

```python
for i in ind:
    print (f'{df.columns[i]} \n')
```

###  Now for KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```
